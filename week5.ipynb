```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

```python
x = np.linspace(0, 10, 10)
y = x**2

plt.plot(x,y, color='g', linestyle='--', marker='o')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafik sin(x)")
plt.show()
```

```python
import pandas as pd

date = pd.date_range(start='2025-10-20', periods=10)
harga_saham = [100, 105, 102, 108, 115, 117, 120, 118, 123, 125]

plt.plot(date, harga_saham, color='g', linestyle='--', marker='o')
plt.xlabel('Tanggal')
plt.ylabel('Harga Saham')
plt.title('Grafik Harga Saham')
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50) * 10

x1 = np.random.rand(50)
y1 = np.random.rand(50) * 10

plt.scatter(x,y, color='r', marker='^', alpha=0.5, label='red')
plt.scatter(x1,y1, color='b', marker='s', alpha=0.5, label='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Grafik Scatter Plot')
plt.show()
```

```python
kategori = ["A", "B", "C", "D"]
jumlah = [10, 20, 30, 40]

plt.bar(kategori, jumlah, color=['g','b','y','orange'])
plt.title('Grafik Pie Chart')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.show()

plt.barh(kategori, jumlah, color=['g','b','y','orange'])
plt.title('Grafik Pie Chart')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.show()
```

```python
kategori = ["A", "B", "C", "D"]
jumlah = [10, 20, 30, 40]

plt.pie(jumlah, labels=kategori, autopct='%1.1f%%', explode=[1, 0, 1, 0])
plt.title('Grafik Pie')
plt.show()
```

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 15, 100)
data3 = np.random.normal(55, 5, 100)

plt.boxplot([data1, data2, data3], tick_labels=['Data 1', 'Data 2', 'Data 3'])
plt.title('Grafik Box Plot')
plt.ylabel('Nilai')
plt.tight_layout()
plt.show()

plt.violinplot([data1, data2, data3], showmedians=True)
plt.title('Grafik Violin Plot')
plt.ylabel('Nilai')
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])
```

```python
data = np.random.rand(200)

plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Grafik Histogram')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Anotasi pada Grafik')

x_peak = np.pi/2
y_peak = np.sin(x_peak)

plt.annotate('Local Maximum', xy=(x_peak, y_peak),
             xytext=(x_peak+1, y_peak+0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
x = [1,2,3,4]
y = [4,2,5,1]

plt.plot(x, y)
plt.title('Anotasi pada Grafik')

plt.annotate('Maximum', xy=(3,5),
             xytext=(3.5, 5.5),
             arrowprops=dict(facecolor='black'))

plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.exp(-x)

fig, ax = plt.subplots(1,2, figsize=(6,4))
ax[0].plot(x,y1, label="sin(x)")
ax[0].set_title("Grafik sin(x)")

ax[1].plot(x,y2, label="cos(x)")
ax[1].set_title("Grafik cos(x)")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(2,2, figsize=(6, 4))
ax[0,0].plot(x,y1, label="sin(x)")
ax[0,0].set_title("Grafik sin(x)")

ax[0,1].plot(x,y2, label="cos(x)")
ax[0,1].set_title("Grafik cos(x)")

ax[1,0].plot(x,y3, label="sin(2x)")
ax[1,0].set_title("Grafik sin(2x)")

ax[1,1].plot(x,y4, label="exp(-x)")
ax[1,1].set_title("Grafik exp(-x)")

plt.tight_layout()
plt.show()
```
