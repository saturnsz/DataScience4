```python
# Import library pandas untuk manipulasi dan analisis data (biasanya digunakan untuk DataFrame)
import pandas as pd
# Import library numpy untuk operasi numerik, terutama pada array dan perhitungan matematika
import numpy as np
# Import matplotlib.pyplot untuk membuat visualisasi data seperti grafik dan plot
import matplotlib.pyplot as plt
# Import seaborn untuk visualisasi data yang lebih menarik dan informatif berbasis matplotlib
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('penguins')
df.info()
```

```python
df = sns.load_dataset('penguins')
df.info()
df['species'].unique()
df['island'].unique()
df.shape
df.head(-5)
```

```python
sns.histplot(df, x='flipper_length_mm', kde=True)
plt.title('Flipper Length')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
sns.histplot(df, x='flipper_length_mm', kde=True)
plt.title('Flipper Length')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
sns.histplot(df, x='flipper_length_mm', kde=True, bins=20)
plt.title('Flipper Length')
plt.show()
```

```python
sns.kdeplot(df, x='flipper_length_mm', fill=True, color='orange')
plt.title('Distribusi Panjang Sirip (mm)')
plt.xlabel('Panjang Sirip (mm)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
sns.kdeplot(df, x='flipper_length_mm',fill=True, color='lightcoral')
sns.rugplot(df, x='flipper_length_mm', color='black')
plt.title('Distribusi Panjang Sirip (mm)')
plt.xlabel('Panjang Sirip (mm)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
sns.scatterplot(df, x='bill_length_mm', y='bill_depth_mm',hue='species')
plt.title('Scatterplot Panjang dan Lebar Baju')
plt.show()
```

```python
q = sns.FacetGrid(df, col='species')
q.map_dataframe(sns.scatterplot, x='bill_length_mm', y='bill_depth_mm')
plt.show()
```

```python
sns.boxplot(df, x='flipper_length_mm', color='Lightgreen')
plt.title('Boxplot Panjang Sirip')
plt.show( )
```

```python
sns.violinplot(df, x='flipper_length_mm', palette='pastel')
plt.title('boxplot panjang sirip')
plt.show()
```

```python
sns.stripplot(df, x='species', y='flipper_length_mm', jitter=0.3, palette='pastel')
plt.title('Boxplot Panjang Sirip')
plt.show()
```

```python
sns.swarmplot(df, x='species', y='flipper_length_mm', hue=True, palette='Set2')
plt.title('Boxplot Panjang Sirip')
plt.xlabel('species')
plt.ylabel('flipper')
plt.show()
```

```python
sns.swarmplot(df, x='species', y='flipper_length_mm', hue=True, palette='Set2')
plt.title('Boxplot Panjang Sirip')
plt.xlabel('species')
plt.ylabel('flipper')
plt.show()
```

```python
sns.regplot(df, x='flipper_length_mm', y='bill_length_mm', color='darkorange')
plt.title('scatterplot panjang dan lebar paruh')
plt.xlabel('panjang sirip (mm)')
plt.xlabel('panjang paruh (mm)')
plt.show()
```

```python
sns.lmplot(df,  x='flipper_length_mm', y='bill_length_mm', hue='species')
plt.title('scatterplot panjang sirip dan panjang paruh')
plt.xlabel('panjang sirip (mm)')
plt.xlabel('panjang paruh (mm)')
plt.show()
```

```python
sns.jointplot(df, x='flipper_length_mm', y='bill_length_mm',
             kind='scatter',color='darkorange')
plt.suptitle('Scatterplot panjang sirip dan panjang paruh', y=1.01)
plt.xlabel('panjang sirip (mm)')
plt.ylabel('panjang paruh (mm)')
plt.show()
```

```python
sns.pairplot(df.dropna(), hue='species')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('heatmap korelasi')
plt.show()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm')
plt.title('heatmap korelasi')
plt.show()
```
